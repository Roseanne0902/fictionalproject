```python
csv_file=open("ibooking.csv", "r")
csv_read=csv.DictReader(csv_file)
line=0
ids= []
dups= []
for row in csv_read:
    if line > 0:
            new= row['Support ID']
            if new not in ids:
                ids.append(new)
            else:
                dups.append(new)   
    line +=1
csv_file.close()

```
